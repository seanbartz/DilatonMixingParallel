```python


def chiral_limit_sigma(tmin,tmax,numtemp,sigma_min,sigma_max,mixing_parameter):

    temps=np.linspace(tmin,tmax,numtemp)

    #light quark mass
    ml=0.1*np.ones(numtemp)

    #chemical potential
    mu=0*np.ones(numtemp)

    lambda1=mixing_parameter*np.ones(numtemp) #parameter for mixing between dilaton and chiral field

    minsigma=sigma_min*np.ones(numtemp)
    maxsigma=sigma_max*np.ones(numtemp)

    a0=0.*np.ones(numtemp)


    tempsArgs=np.array([temps,mu,ml,minsigma,maxsigma,a0,lambda1]).T

    "Create a pool that uses all available cpus"
    processes_count=os.cpu_count()    
    processes_pool = Pool(processes_count)

    truesigma=get_all_sigmas_parallel(allSigmas,tempsArgs,processes_pool)
    truesigma=np.array(truesigma)
    processes_pool.close()

    return temps,truesigma,max(truesigma[:,0])

```

```python
color_map = plt.colormaps.get_cmap('plasma')

lambda1_styles = {
    '5.3': '-', 
    '5.4': '--', 
    '6.0': '-.', 
    '5.6': ':', 
    '6.5': (0, (3, 5, 1, 5)), 
    '7.0': (0, (3, 1, 1, 1))
}

lambda1_values = np.array([7, 6.5, 6, 5.6, 5.4, 5.3])
colors = color_map(np.linspace(0.8, 0, len(lambda1_values)))

for i, lambda1 in enumerate(lambda1_values):
    linestyle = lambda1_styles[f'{lambda1:.1f}']
    color = colors[i]
    temps_name = f'temps{str(lambda1).replace(".", "")}'
    sigma_name = f'sigma{str(lambda1).replace(".", "")}'
    temps = eval(temps_name)
    sigma = eval(sigma_name)
    plt.plot(temps, sigma[:,0]**3/1e9, label=f'$\lambda_1={lambda1}$', color=color, linestyle=linestyle)
    plt.plot(temps, -sigma[:,0]**3/1e9, color=color, linestyle=linestyle)

plt.legend()
plt.xlabel('Temperature (MeV)')
plt.ylabel('$\sigma$ (GeV$^3$)')
#make the figure look nice
plt.tight_layout()
#make the x axis go up to 140
plt.xlim(12, 145)
#save the figure as an eps file
plt.savefig('chiral_limit_2_flavor_sigma_T.eps')
plt.show()
```
