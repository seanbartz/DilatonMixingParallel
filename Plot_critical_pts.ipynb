```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.interpolate import UnivariateSpline
import numpy as np
# Load the updated data from the provided file
data_updated = pd.read_csv('data/critical_points.csv')

# Sort the data by 'ml' and 'mu_crit' to correctly join the points
data_sorted = data_updated.sort_values(['ml', 'mu_crit'])
# Remove the rows where mu_crit is greater than 1000
data_cleaned = data_sorted[data_sorted['mu_crit'] <= 1000]

# Create a scatter plot and line plot on top for each 'ml' value
plt.figure(figsize=(10,6))

for ml in data_cleaned['ml'].unique():
    subset = data_cleaned[data_cleaned['ml'] == ml]
    sns.scatterplot(x='mu_crit', y='Tc_crit', hue='lambda1', palette="viridis", marker='o', data=subset, label=f'ml={ml}')
    plt.plot(subset['mu_crit'], subset['Tc_crit'], label=None)

plt.xlabel('mu_crit (Chemical Potential)')
plt.ylabel('Tc_crit (Critical Temperature)')
plt.title('Critical Temperature vs Chemical Potential (Color-coded by Lambda) - Cleaned Data')
plt.legend(title='ml')
plt.show()
```
